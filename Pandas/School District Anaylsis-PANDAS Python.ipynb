```python
import pandas as pd
```

```python
schools_pd = "schools_complete.csv"
students_pd = "students_complete.csv"

schools_df = pd.read_csv(schools_pd, encoding="iso-8859-1", low_memory=False)
students_df = pd.read_csv(students_pd, encoding="iso-8859-1", low_memory=False)
                                   
schools_df.head()
students_df.head()
```

```python
total_schools = schools_df["School ID"].count()
total_schools

total_students = schools_df["size"].sum()
total_students

total_budget = schools_df["budget"].sum()
total_budget

average_math_score = students_df["math_score"].mean()
average_math_score

average_reading_score = students_df["reading_score"].mean()
average_reading_score

Number_pass_math = students_df["math_score"].loc[students_df["math_score"] > 70].count()
percent_pass_math = (Number_pass_math/total_students)*100
percent_pass_math

Number_pass_reading = students_df["reading_score"].loc[students_df["reading_score"] > 70].count()
percent_pass_reading = (Number_pass_reading/total_students)*100
percent_pass_reading

Overall_pass_rate = (percent_pass_math + percent_pass_reading)/2
Overall_pass_rate
```

```python
summary_data = pd.DataFrame({
    "Total School":[total_schools],
    "Total Number of Students":[total_students],
    "Total Budget":[total_budget],
    "Average Reading Score":[average_reading_score],
    "Average Math Score":[average_math_score],
    "Reading Score Percentage":[percent_pass_math],
    "Math Score Percentage":[percent_pass_reading],
})

summary_data
```

```python
school_type = school_summary["type"].unique().str.get(0)
total_school_budget = school_summary["budget"].unique().str.get(0)
total_school_students = school_summary.count()["Student ID"]
per_student_budget = total_school_budget/total_school_students
```

```python
school_summary_df = pd.DataFrame({"School Type": school_type,
                                 "Total_Students": total_school_students,
                                 "Total Budget": total_school_budget,
                                 "Per Student Budget": per_student_budget,
                                 "Percent Passing Math": percent_passing_math_school,
                                 "Percent Passing Reading": percent_passing_reading_school,
                                 "Overall Passing Rate": percent_overall_passing_school})
school_summary_df
```
